## Popularity Based Recommendations

```python
new_df = rating.merge(books,on='Book',how='left')
```

- new_df will have all the columns from both rating and books dataframes, with the rows aligned based on the matching Book values. This allows you to combine information from both dataframes into a single dataframe, where you can perform further analysis or visualization.

```python
new_df.rename({"ID":"User_id",'Id':'Book_id'},inplace = True,axis = 1)
```

```python
new_df.shape
```

```python
new_df
```

```python
## Taking books and num of rating perform on that Books
```

```python
num_rating_df = new_df.groupby('Book').count()['User Rating'].reset_index()
num_rating_df.rename({'User Rating':"num_rating"},inplace=True,axis = 1)
```

- num_rating_df provides a summary of the number of ratings for each book in the merged dataframe new_df. This information can be useful for analyzing the popularity of different books and for identifying which books have more user ratings in the dataset.

```python
num_rating_df
```

```python
## Similarly we are find average rating of Book
```

```python
avg_rating_df = new_df.groupby('Book').mean()['User Rating'].reset_index()
avg_rating_df.rename({'User Rating':"avg_rating"},inplace=True,axis = 1)
avg_rating_df
```

- avg_rating_df provides a summary of the average rating for each book in the merged dataframe new_df. This information can be useful for analyzing the overall quality or popularity of different books, and for identifying which books have a higher average rating in the dataset. It can also be used to compare the ratings of different books against each other.

## Merging two values num_rating_df and avg_rating_df on top of Book-Title

```python
popular_df = num_rating_df.merge(avg_rating_df,on = 'Book')
```

- num_rating_df contains the count of ratings for each book, while avg_rating_df contains the average rating for each book. By merging these two dataframes on the Book column, popular_df combines the information about the number of ratings and the average rating for each book in a single dataframe.
- popular_df provides a summary of the popularity of different books based on the number of ratings and the average rating. This information can be useful for identifying which books are both popular and well-rated, or which books are popular despite having a lower average rating. The merged dataframe popular_df allows for easy comparison of these two metrics across all books in the dataset.

```python
popular_df
```

## collabarative Filtering based recommender System

## taking those users those who done a more than 200 ratings

```python
x = new_df.groupby('User_id').count()['Book'] > 200
Good_Reader = x[x].index
```

- The resulting output is a Pandas Series containing a boolean value for each User_id, indicating whether they have rated more than 200 books or not. The boolean value is based on the comparison > 200 which returns True if the user has rated more than 200 books and False otherwise.
- Therefore, the resulting x Series can be used to identify users who have rated more than 200 books in the dataset. This information can be useful for analyzing the behavior of power users or heavy readers in the dataset, and for making recommendations to them based on their past reading preferences.

```python
Good_Reader
```

## Taking only those column those who presents in Good_Reader

```python
filter_rating = new_df[new_df['User_id'].isin(Good_Reader)]
```

```python
filter_rating
```

- filter_rating contains only the ratings of the users who have rated more than 200 books in the dataset. This filtered dataframe can be used for further analysis of the reading behavior and preferences of these high-volume readers, and for making personalized recommendations to them based on their past ratings.

```python
more_than_50_rate = filter_rating.groupby('Book').count()['User Rating'] >=50
famous_books = more_than_50_rate[more_than_50_rate].index
```

- The resulting output is a Pandas Series containing boolean values for each book, indicating whether it has received 50 or more ratings from the Good_Reader users or not. The boolean value is based on the comparison >= 50, which returns True if the book has received 50 or more ratings and False otherwise
- more_than_50_rate Series can be used to identify the books that have received 50 or more ratings from the high-volume Good_Reader users. This information can be useful for identifying the most popular and widely-read books among these users, and for making recommendations to them based on their reading preferences.

```python
famous_books.values
```

```python
final_rating = filter_rating[filter_rating['Book'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index = 'Book',columns = 'User_id',values = 'User Rating')
```

```python
pt.fillna(0,inplace = True)
pt
```

## now we finding a eulidean distance between the books

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python

# Creating Recommendation Function
def recommend(BookName):
    try:
        # index fetch
        index = np.where(pt.index == BookName)[0][0]
#         index = np.where(pt.index == 'The Martian')[0][0]
        similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]

        data = []
        for i in similar_items:
            item = []
            temp_df = new_df[new_df['Book'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book')['Book'].values))
            item.extend(list(temp_df.drop_duplicates('Book')['Authors'].values))
            data.append(item)

        return data

    except:
        print("Book Not Found in Database")
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(matrix)
```

```python
def get_recommends(book = ""):
  x=pt.loc[book].array.reshape(1, -1)
  distances,indices=model_knn.kneighbors(x,n_neighbors=6)
  R_books=[]
  for distance,indice in zip(distances[0],indices[0]):
    if distance!=0:
      R_book=pt.index[indice]
      R_books.append([R_book,distance])
  recommended_books=[book,R_books[::-1]]
  return recommended_books
```
